```python
import pandas as pd

df = pd.read_json('datasets/2022-2023/team_stats.json').transpose()

dep = "regulation_wins"
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

fig, axs = plt.subplots(1, 2, figsize=(11, 5))
sns.scatterplot(data=df, y=dep, x="shots_per_game", ax=axs[0]).set(title="Shots vs Wins (2022-2023)")
sns.scatterplot(data=df, y=dep, x="shots_against_per_game", ax=axs[1]).set(title="Shots against vs Wins (2022-2023)")
for ax in axs:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```
